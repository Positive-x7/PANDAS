```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('GoogleApps.csv')
```

```python
s = pd.Series(data = [10, 5, 15, 20, 10],
              index = [1, 2, 3, 4, 5])
s.plot()
plt.show()
```

```python
df['Category'].value_counts().head().plot(kind = 'pie')
```

```python
df['Category'].value_counts().plot(kind = 'barh', figsize = (10 , 5) , grid=True)
```

```python
d1 = df[df['Type'] == 'Free'].groupby('Content Rating') ['Installs'].mean() 
d1.plot(kind= 'barh') 
```

```python
d2 = df[df['Type'] == 'Paid'].groupby ('Content Rating') ['Installs'].mean()
d2.plot(kind = 'barh')
```

```python
d = df.pivot_table(index = 'Content Rating',
                   columns = 'Type',
                   values = 'Installs',
                   aggfunc = 'mean')

d.plot(kind = 'barh', subplots = True, layout = (1,2), sharey = True)
d.plot(kind = 'barh', grid = True)
```
